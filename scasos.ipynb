````python
import numpy as np
import matplotlib.pyplot as plt

#loc: centro
#scale: desviación estándar
#size: número de muestras
#density: para poder graficar varias

x1 = np.random.normal(loc=0, scale=1, size=10000)
x2 = np.random.normal(loc=2, scale=2, size=10000)

#print(x1)

plt.hist(x1, bins=50, alpha=.5, density=True, label="x1")
plt.hist(x2, bins=30, alpha=.3, density=True, label="x2")

plt.grid();
plt.legend();
````
